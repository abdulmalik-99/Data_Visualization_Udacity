## Investigation Overview


In this exploration, I am trying to identify the most important featurse that are  related to the Ford Bike Sharing system, and this data covers the Greater San Francisco Bay Area in February 2019. the key insights are:

- the most of users are subscribers

- Friday and Thursday less than other day in terms of users 

- The Most trips occur during the working hours of 6am to 9am and 4pm to 6pm



## Dataset Overview

> Ford GoBike is a bike-sharing system in the San Francisco Bay Area, operated by Motivate International. The system allows users to rent bicycles for short trips, using a network of automated kiosks and mobile app. Users can purchase a membership or pay for each ride individually. The system provides data on bike availability, usage, and trip information to the public, which can be used for transportation planning, research, and other purposes.





```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df=pd.read_csv("201902-fordgobike-tripdata-clean.csv")
```

## (Visualization 1)

> The plot shows the most of users are subscribers


```python
plt.figure(figsize=[14,7])
fig=sns.barplot(x=df['user_type'].value_counts().index,y=df['user_type'].value_counts(),palette='mako')
fig.set(ylabel='Frequency',xlabel='User Type',title='Distribution of User Type');
```

## (Visualization 2)

>  The plot shows Fridays and Thursdays have fewer users compared to other days.

```python
plt.figure(figsize=[14,7])

fig = sns.barplot(x=df['days_ordered'].value_counts().index, y= df['day'].value_counts(),palette='mako')
fig.set(xlabel='day', ylabel='Frequency',title='Number of trips each day' );
```

## (Visualization 3)



The plot shows a low number of trips during night hours (10pm to 5am) due to sleep and rest. The peak times for trips are from 6am to 9am and 4pm to 6pm as these are working hours.

```python
plt.figure(figsize=[14,7])
fig = sns.barplot(x=df['hour'].value_counts().index, y= df['hour'].value_counts(),palette='mako')
fig.set(xlabel='Hour', ylabel='Frequency',title='Number of trips each Hour' );
```
